```python
import cv2 as cv
import numpy as np
```

```python
frameW = 1280
frameH = 720

cap = cv.VideoCapture(0)
 
if not cap.isOpened():
    print("Could not open camera")
    exit()
cap.set(3,frameW)
cap.set(4,frameH)
cap.set(10,150)

while True:
    ret,img = cap.read()
    
    if not ret:
        print("Could not read frame")
        break
    cv.imshow("video", img)
   
    
    if cv.waitKey(1) ==  ord('q'):
        break
        
cv.destroyAllWindows()
```

```python
img  = cv.imread('res/cards.jpg')

w,h = 250,350
pts1 = np.float32([[111,219],[287,188],[154,482],[352,440]])
pts2 = np.float32([[0,0],[w,0],[0,h],[w,h]])
matrix = cv.getPerspectiveTransform(pts1,pts2)
imgO = cv.warpPerspective(img,matrix,(w,h))

cv.imshow("open",img)
cv.imshow("trans", imgO)

cv.waitKey(0)
```
